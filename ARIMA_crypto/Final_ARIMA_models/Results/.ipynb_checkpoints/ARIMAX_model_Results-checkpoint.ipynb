```python
import pandas as pd 
train = pd.read_csv('training_residuals_nodiff.csv')
test = pd.read_csv('test_residuals_nodiff.csv')
```

```python
# Test Residuals 
for coin in test['Coin'].unique():
    coin_data = test[test['Coin'] == coin]
    plt.figure(figsize=(12, 6))
    plt.plot(coin_data['Date'], coin_data['Residual'], color='red')
    plt.title(f'Test Residuals {coin}')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.show()
```

```python
#Training Actual v. Predicted
for coin in train['Coin'].unique():
    coin_data = train[train['Coin'] == coin]
    plt.figure(figsize=(12, 6))
    plt.plot(coin_data['Date'], coin_data['actual'], label='actual', color='blue')
    plt.plot(coin_data['Date'], coin_data['predicted'], label='Predicted', color='red')
    plt.title(f'Training Actual vs Predicted Values for {coin}')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.legend()
    plt.show()
```
