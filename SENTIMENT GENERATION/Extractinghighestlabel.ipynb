**EXTRACTING HIGHEST PROBABILITY FROM THE ORDERED DICT**

```python
# Define a function to get the highest probability label
def get_highest_prob_label(prob_ordered_dict_str):
    # Convert the string representation of the ordered dictionary back to an ordered dictionary
    prob_ordered_dict = eval(prob_ordered_dict_str)
    # The first item in the ordered dictionary will be the label with the highest probability
    highest_prob_label = next(iter(prob_ordered_dict))
    return highest_prob_label

# Apply the function to the 'sentiment_probabilities' column to create a new 'highest_prob_label' column
df['highest_prob_label'] = df['sentiment_probabilities'].apply(get_highest_prob_label)
```
